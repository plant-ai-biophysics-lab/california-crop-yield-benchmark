```python
import os 
import xarray 
import numpy
import matplotlib.pyplot as plt


import os
import numpy as np
import rasterio
import matplotlib.pyplot as plt

def plot_landsat5_all_bands_by_index(folder_path, month):
    """
    Plot all six Landsat 5 bands (Blue, Green, Red, NIR, SWIR1, SWIR2)
    separately in a 2-row by 3-column layout.

    Args:
        folder_path (str): Path to the folder containing .tif files.
        month (int): Index of the file to read.

    Returns:
        None. Displays the 6-band images.
    """
    
    def list_tif_files(folder_path):
        return sorted([f for f in os.listdir(folder_path) if f.endswith('.tif')])

    def read_tif_by_index(folder_path, month):
        files = list_tif_files(folder_path)
        if month < 0 or month >= len(files):
            raise IndexError("Month index out of range.")
        file_path = os.path.join(folder_path, files[month])
        print(f"Image name: {files[month]}")
        with rasterio.open(file_path) as src:
            image = src.read()
        return image  # shape: (bands, height, width)

    def normalize(array):
        """
        Normalize the array to [0, 1] for visualization after handling NaN values.
        """
        array = np.where(np.isnan(array), np.nanmin(array), array)
        norm_array = (array - array.min()) / (array.max() - array.min())
        return norm_array

    image = read_tif_by_index(folder_path, month)

    # Landsat 5 bands: [1: Blue, 2: Green, 3: Red, 4: NIR, 5: SWIR1, 7: SWIR2]
    band_names = ["Blue", "Green", "Red", "NIR", "SWIR1", "SWIR2"]

    if image.shape[0] < 6:
        raise ValueError(f"Expected at least 6 bands, but found {image.shape[0]}")

    bands = [normalize(image[i, :, :]) for i in range(6)]

    # Plotting
    fig, axes = plt.subplots(2, 3, figsize=(20, 12), facecolor="white")

    for ax, band, name in zip(axes.flat, bands, band_names):
        ax.set_facecolor("white")
        im = ax.imshow(band, cmap="gray", vmin=0, vmax=1)
        ax.set_title(name, fontsize=12, color="black")
        ax.axis("off")

    plt.suptitle("Landsat 5 Individual Bands", fontsize=14, fontweight="bold", color="black")
    plt.tight_layout()
    plt.show()
```

```python
plot_landsat5_all_bands_by_index(folder_path = '../data/counties/SanFrancisco/data/landsat/2018', month = 5)
```
