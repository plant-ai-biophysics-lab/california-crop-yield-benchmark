```python
import rasterio
import matplotlib.pyplot as plt

def plot_tif(tif_path, title=None, cmap="viridis"):
    with rasterio.open(tif_path) as src:
        data = src.read(1)  # read the first band
        plt.figure(figsize=(8, 6))
        plt.imshow(data, cmap=cmap)
        plt.colorbar(label='Value')
        if title:
            plt.title(title)
        plt.axis('off')
        plt.show()
```

```python
plot_tif("../data/counties/Napa/data/cdl/2019/Napa_CDL_2019_cropped.TIF", title=None, cmap="viridis")
```
